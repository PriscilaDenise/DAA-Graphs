Imported necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

Defined functions that represent the time complexity of each sorting algorithm.

```python
def quick_sort_complexity(n):
    return n * np.log2(n)

def merge_sort_complexity(n):
    return n * np.log2(n)

def insertion_sort_complexity(n):
    return n**2
```

Generate data for plotting 

```python
n_values = np.linspace(1, 10, 10)  # 1 to 1000 input sizes
quick_sort_times = quick_sort_complexity(n_values)
merge_sort_times = merge_sort_complexity(n_values)
insertion_sort_times = insertion_sort_complexity(n_values)
```

Use Matplotlib to create the graph. Plot each function, label the axes, and add a title and legend.

```python
plt.figure(figsize=(10, 6))
plt.plot(n_values, quick_sort_times, label='Quick Sort (O(n log n))', color='blue')
plt.plot(n_values, merge_sort_times, label='Merge Sort (O(n log n))', color='green')
plt.plot(n_values, insertion_sort_times, label='Insertion Sort (O(n^2))', color='red')
plt.yscale('log')  # Optional: Use logarithmic scale for better visualization

# Labeling the axes and the plot
plt.xlabel('Input Size (n)')
plt.ylabel('Time Complexity')
plt.title('Comparison of Sorting Algorithm Time Complexities')
plt.legend()
plt.grid(True)
plt.show()
```
